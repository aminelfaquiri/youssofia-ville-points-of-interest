```python
import folium
import json
from geopy.distance import geodesic
import openrouteservice as ors
```

### create map with speciale place :

```python
# Fonction to select the route from two places :

# OpenRouteService API key :
API_KEY = '5b3ce3597851110001cf6248bfe7477f80bb48d8b0dbafec4a988159'

# Function to generate the route between two locations
def generate_route(start_coords, end_coords, mode='driving-car'):
    client = ors.Client(key=API_KEY)
    route = client.directions(
        coordinates=[start_coords, end_coords],
        profile=mode,
        format='geojson'
    )
    return route

def calculate_distance(point1, point2):
    return geodesic(point1, point2).meters
```

```python
def centre_de_ville(locations):
    sum_lat = 0
    sum_lon = 0
    for location in locations:
        sum_lat += location[0]
        sum_lon += location[1]
    return round(sum_lat / len(locations), 5), round(sum_lon / len(locations), 5)
    
# Lire les données à partir du fichier JSON
with open('yousoufian.json') as f:
    data = json.load(f)

# Coordonnées des lieux
locations = []
for feature in data['features']:
    locations.append(feature['geometry']['coordinates'][::-1])

# Créer la carte :
city_coordinates = centre_de_ville(locations)
city_map = folium.Map(location=city_coordinates, zoom_start=15)

# Définir les couleurs d'icône pour différents types
icon_colors = {
    "transport": "blue",
    "cafe": "green",
    "pharmacie": "red",
    "sport place": "yellow",
    "ecole": "violet",
    "Hôpital": "orange"
}

# Liste des icônes
icons = {
    "ecole": "book",
    "cafe": "music", 
    "pharmacie": "home",
    "terran": "star",
}

# Ajouter des marqueurs pour chaque élément dans les données JSON
for feature in data['features']:
    properties = feature['properties']
    name = properties.get('name', 'Unknown')
    description = properties.get('description', 'No description')
    location = feature['geometry']['coordinates'][::-1]
    marker_type = properties.get('type')
    
    # Créer une icône personnalisée pour chaque type :
    icon_color = icon_colors.get(marker_type, "gray")
    icon_icon = icons.get(marker_type, "info-sign")
    icon = folium.Icon(color=icon_color, icon= icon_icon)

    # Créer une infobulle avec les informations du marqueur
    popup = f"<b>{name}</b><br>{description}"

    # Ajouter le marqueur à la carte
    folium.Marker(location=location, popup=popup, icon=icon).add_to(city_map)

##############################################################################
# Example coordinates for start and end locations :
start_coords = (-8.521736745223308, 32.247578759898246)
end_coords = ( -8.522141367867079, 32.24861094447948)

route = generate_route(start_coords, end_coords, mode='driving-car')

def style_function(feature):
    return {
        'color': 'red', 
        'weight': 10,
        'opacity': 0.7
    }

route_geojson = folium.GeoJson(route, name='Youcode_to_hospital', style_function=style_function)

city_map.add_child(route_geojson)

folium.Marker(location=start_coords, popup='Start ').add_to(city_map)
folium.Marker(location=end_coords, popup='End').add_to(city_map)
##############################################################################
start_coords2 = (-8.532431728054434, 32.248983104488474)
end_coords2 = (-8.521736745223308, 32.247578759898246)
# Generate the route using OpenRouteService :
route2 = generate_route(start_coords2,end_coords2 , mode='driving-car')
# Function to customize the style of the route
def style_function2(feature):
    return {
        'color': 'black', 
        'weight': 10,
        'opacity': 0.7
    }
# Add the route to the map :
route_geojson2 = folium.GeoJson(route2, name='test', style_function=style_function2)
city_map.add_child(route_geojson2)
# Add markers for start and end locations :
folium.Marker(location=start_coords2, popup='Start 2').add_to(city_map)
folium.Marker(location=end_coords2, popup='End 2').add_to(city_map)
####################################################################################

# Add a layer control to the map :
folium.LayerControl().add_to(city_map)
# Display the map :
city_map
```
