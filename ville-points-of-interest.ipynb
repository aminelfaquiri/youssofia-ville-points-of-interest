### change osm to df :

```python
import pandas as pd
import xml.etree.ElementTree as ET

# Replace 'path/to/osm_file.xml' with the actual path to your OSM XML file
osm_file_path = 'Youssoufia.osm'

# Read the OSM XML data from the file :
with open(osm_file_path, 'r', encoding='utf-8') as file:
    osm_data = file.read()

# Parse the XML data
tree = ET.ElementTree(ET.fromstring(osm_data))
root = tree.getroot()

# Extract node information
nodes_data = []
for node in root.findall('.//node'):
    node_data = {
        # 'id': int(node.attrib['id']),
        # 'visible': node.attrib['visible'],
        # 'version': int(node.attrib['version']),
        # 'changeset': int(node.attrib['changeset']),
        # 'timestamp': node.attrib['timestamp'],
        # 'user': node.attrib['user'],
        # 'uid': int(node.attrib['uid']),
        'lat': float(node.attrib['lat']),
        'lon': float(node.attrib['lon']),
    }
    nodes_data.append(node_data)
# Create a pandas DataFrame from the extracted information
df = pd.DataFrame(nodes_data)
df

my_all_placement = nodes_data
my_all_placement
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="geoapiExercises")
def city_state_country(coord):
    location = geolocator.reverse(coord, exactly_one=True)
    address = location.raw['address']
    return address
print(city_state_country("32.240434, -8.529354"))	
```

```python
for i in my_all_placement :
    print(i)
    # add value in this dictionary :
    name = city_state_country(f"{str(i['lat'])}, {str(i['lon'])}")
    print(name)
    i['info'] = name
```

### Use folium :

```python
def centre_de_ville(locations):
    sum_lat = 0
    sum_lon = 0
    for location in locations:
        sum_lat += location[0]
        sum_lon += location[1]
    return round(sum_lat / len(locations), 5), round(sum_lon / len(locations), 5)
    
# Lire les données à partir du fichier JSON
with open('yousoufian.json') as f:
    data = json.load(f)

# Coordonnées des lieux
locations = []
for feature in data['features']:
    locations.append(feature['geometry']['coordinates'][::-1])

# Créer la carte
city_coordinates = centre_de_ville(locations)
city_map = folium.Map(location=city_coordinates, zoom_start=15)

# Définir les couleurs d'icône pour différents types
icon_colors = {
    "transport": "blue",
    "cafe": "green",
    "pharmacie": "red",
    "sport place": "yellow",
    "ecole": "violet",
    "Hôpital": "orange"
}

# Liste des icônes
icons = {
    "ecole": "book",
    "cafe": "music", 
    "pharmacie": "home",
    "terran": "star",
}

# Ajouter des marqueurs pour chaque élément dans les données JSON
for feature in data['features']:
    properties = feature['properties']
    name = properties.get('name', 'Unknown')
    description = properties.get('description', 'No description')
    location = feature['geometry']['coordinates'][::-1]
    marker_type = properties.get('type')
    
    # Créer une icône personnalisée pour chaque type
    icon_color = icon_colors.get(marker_type, "gray")
    icon_icon = icons.get(marker_type, "info-sign")
    icon = folium.Icon(color=icon_color, icon= icon_icon)

    # Créer une infobulle avec les informations du marqueur
    popup = f"<b>{name}</b><br>{description}"

    # Ajouter le marqueur à la carte
    folium.Marker(location=location, popup=popup, icon=icon).add_to(city_map)

# Afficher la carte :
city_map
```
